```python

H = 1
sigma = 0

time = np.arange(-10.1, 10.1, 0.1)

h_list = [0 if t < 0 else H for t in time]

plt.figure(figsize=(10, 5))

for i in range(10):
    sigma += 0.5
    gaussian_deriv = []
    for t in time:
        gaussian_deriv.append(
            -t
            * np.exp(((-(t ** 2)) / (2 * sigma ** 2)))
            / (np.sqrt(2 * np.pi) * sigma ** 3)
        )
    
    convolved = np.convolve(h_list, gaussian_deriv)
    plt.plot(convolved)
    

pad_len = int(0.5 * (convolved.shape[0] - len(h_list)))
plt.plot(np.pad(h_list, (pad_len, pad_len)))

plt.xlim(130, 280)
plt.title("Convolution of Step Edge with Gaussian derivative")

plt.legend()
plt.show()
    
```
